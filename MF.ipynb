## About the data

The quarterly  financial analysts' forecast of earnings per share (EPS) data is collected from
I/B/E/S, Thomson Reuters. We consider the time from quarter 1, 2000 to quarter 4, 2016. We perform some filtering procedures
1. we only select  firms that appear at least 48 quarters out of the total 68 quarters, which allows us to have sufficient historical records. 
2. we remove  rms with less than ten di erent analysts during the sample period. 
3. we remove analysts who make less than three forecasts in the considered period. 
The  final sample involves 2082 firms and 9785 analysts.

Beacuese this data is propreitory we provided the code with example for only one firm 'Apple Inc' (AAPL). The full data can be accessed from here with a subscription https://wrds-www.wharton.upenn.edu/pages/get-data/ibes-thomson-reuters/ 



```python
# read data from the file and create analyst forecast data (AAP_X) and actual eps value (AAP_Y)

AAPL = pd.read_csv('Data/AAPL.csv')
AAP_X = AAPL[[ 'yq','firmID','analystID','value']]
AAP_Y = AAPL[[ 'yq','firmID','actual']]
AAP_Y = AAP_Y.groupby(['yq', 'firmID']).first()
A_X = AAP_X.set_index(['yq', 'firmID', 'analystID']).unstack()
A_X = A_X.reset_index().drop('firmID', axis=1).set_index('yq')
A_X.columns = list(A_X.columns.droplevel())
```

### Compare Imputation Performance with Additional Missing data

```python
## Calculate Initial Percentate of Missing Value in the data set
print("Intial percentage of missing value", np.round(per_miss(A_X)*100,2),'%')
```

```python
#Create 80% missing value by randomly removing 21% of the existing data
Extra_Missing = Creat_missing(A_X, 0.21)
```

```python
np.round(per_miss(Extra_Missing)*100,2)
```

```python
#Zero Imputation
print("Zero Imputation Performance")
Zero_filled = Zero_impute(Extra_Missing)
Evaluate_performance(Zero_filled, A_X)
```

```python
#Random Imputation
print("Random Imputation Performance")
random_filled = random_walk(Extra_Missing, AAP_Y )
Evaluate_performance(random_filled,  A_X.drop(A_X.tail(1).index))
```

```python
#k_NN Imputation
print("k-NN Imputation Performance")
kNN_filled, _ = KNN(k=5).fit_transform(Extra_Missing.values)
Evaluate_performance(kNN_filled, A_X)
```

```python
#Mean Imputation
print("Mean Imputation Performance")

mean_filled, _ = SimpleFill("mean").fit_transform(Extra_Missing.values )
Evaluate_performance(mean_filled, A_X)
```

```python
#Perform Matrix (MF) Imputation
#the two output are filled values (MF_imputed) and the Model(the tensorflow model for MF with learned parameters)

MF = MatrixFactorization(rank = 5,  min_improvement=0.0001)
MF_filled, Model = MF.fit_transform(Extra_Missing.values)
```

### Finally The Imputation Performance of Proposed MF model

```python
print("Matrix Factorization (MF) Imputation Performance")
Evaluate_performance(MF_filled, A_X)
```

### Evalaute the Learned Factors

```python
#Perform Matrix Factorization on whole data

MF = MatrixFactorization(rank = 5,  min_improvement=0.0001)
MF_filled, Model = MF.fit_transform(A_X.values)
```

```python
MF_imputed = pd.DataFrame(MF_filled,columns= A_X.columns, index= A_X.index)
```

### The larned U and V, and thier similarity comparison

```python
q = Model.weights[0].numpy()
q= np.round(cosine_similarity(q, q),3)
angular_dis_Q = np.arccos(q) / np.pi

quarters = list(A_X.reset_index().groupby(['yq']).first().reset_index()['yq'])
angular_dis_Q =pd.DataFrame(angular_dis_Q,columns= quarters, index= quarters)

#a=pd.DataFrame(angular_dis)
ax = sns.heatmap(angular_dis_Q, linewidth=0.00001,cbar_kws={'label': 'Cosine Distance'})
#plt.xlabel("Quarters")
plt.tight_layout()
#plt.savefig('Heatmap_Quarter_apple_2.png', dpi= 500)
plt.show()
```

```python
a = Model.weights[1].numpy()
#b = b[:20,:]
a= np.round(cosine_similarity(a, a),3)
angular_dis_A = np.arccos(a) / np.pi
analyst_name = list(A_X.columns)
analyst_latent = pd.DataFrame(angular_dis_A, index =analyst_name, columns = analyst_name)

analyst_latent.sort_values(by = analyst_name, inplace = True)
analyst_latent.sort_values(by = analyst_name, axis = 1,  inplace = True)
ax = sns.heatmap(analyst_latent, linewidth=0.000001,cbar_kws={'label': 'Cosine Similarity'})
plt.xlabel("Analyst")
plt.ylabel("Analyst")
plt.tight_layout()
#plt.savefig('Heatmap_analyst.png', dpi= 500)
plt.show()
```
